```python
from PIL import Image
import numpy as np
from IPython.display import display
from pynq import Xlnk
from pynq import Overlay

import matplotlib.pyplot as plt 

from pynq import allocate
import signal
from time import sleep
```

```python
hls4ml_design = Overlay(
    "/home/xilinx/jupyter_notebooks/SAVE/quantized_312.bit")
```

```python
dma = hls4ml_design.hier_0.axi_dma_0
```

```python
xlnk = Xlnk()

in_buffer = xlnk.cma_array(shape=(512, 16, 1), 
                           dtype=np.float32, cacheable=1)
out_buffer = xlnk.cma_array(shape=(3, 1), 
                            dtype=np.float32, cacheable=1)
```

```python
def run_kernel():
    dma.sendchannel.transfer(in_buffer)
    dma.recvchannel.transfer(out_buffer)
    dma.sendchannel.wait()
    dma.recvchannel.wait()
```

Az alábbi cellát el kell indítani, majd körülbelül 4 másodperc után a stop gombra kell kattintani. "KeyboardInterrupt: " üzenettel leáll, ezek után az ez utáni cellákkal sorrendben kell tovább menni.

```python
in_buffer.flush()
out_buffer.flush()

dma.sendchannel._clear_interrupt()
dma.recvchannel._clear_interrupt()
run_kernel()
```

```python
in_buffer.flush()
out_buffer.flush()

dma.sendchannel._active_buffer.invalidate()
dma.recvchannel._active_buffer.invalidate()
dma.sendchannel._clear_interrupt()
dma.recvchannel._clear_interrupt()

run_kernel()
```
